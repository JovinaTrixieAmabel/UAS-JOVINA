```python
# Pertanyaan 1:
# Buat scatterplot untuk melihat hubungan antara jumlah produk yang diproduksi (Production) dengan jumlah produk yang terjual (Sales). Sertakan label untuk sumbu x dan y.

import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Month": ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"],
    "Production": [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    "Sales": [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    "Inventory": [100, 200, 300, 200, 100, 200, 100, 200, 300, 300, 200, 100],
    "Returns": [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df["Production"], df["Sales"], color='blue')
plt.title("Hubungan antara Produksi dan Penjualan")
plt.xlabel("Produksi")
plt.ylabel("Penjualan")
plt.grid(True)
plt.show()
```

```python
# Pertanyaan 2:  
# Buat histogram untuk menggambarkan distribusi jumlah produk yang dikembalikan (Returns). Sertakan judul dan label sumbu.

import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Month": ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"],
    "Production": [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    "Sales": [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    "Inventory": [100, 200, 300, 200, 100, 200, 100, 200, 300, 300, 200, 100],
    "Returns": [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

# Create histogram for Returns
plt.figure(figsize=(10, 6))
plt.hist(df["Returns"], bins=10, color='green', edgecolor='black')
plt.title("Distribusi Jumlah Produk yang Dikembalikan")
plt.xlabel("Jumlah Produk yang Dikembalikan")
plt.ylabel("Frekuensi")
plt.grid(True)
plt.show()
```

```python
# Pertanyaan 3: 
# Misalkan bulan-bulan yang jumlah produksinya di atas rata-rata disebut "High Production", bulan-bulan yang jumlah 
# penjualannya di atas rata-rata disebut "High Sales", dan bulan-bulan yang jumlah persediaannya di atas rata-rata disebut "High Inventory". 
# Buat diagram Venn untuk menggambarkan hubungan antara tiga kelompok bulan ini.

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Data
data = {
    "Month": ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"],
    "Production": [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    "Sales": [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    "Inventory": [100, 200, 300, 200, 100, 200, 100, 200, 300, 300, 200, 100],
    "Returns": [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

# Calculate averages
average_production = df["Production"].mean()
average_sales = df["Sales"].mean()
average_inventory = df["Inventory"].mean()

# Determine months in each category
high_production_months = set(df[df["Production"] > average_production]["Month"])
high_sales_months = set(df[df["Sales"] > average_sales]["Month"])
high_inventory_months = set(df[df["Inventory"] > average_inventory]["Month"])

# Create Venn diagram
plt.figure(figsize=(10, 8))
venn3([high_production_months, high_sales_months, high_inventory_months], 
      ('High Production', 'High Sales', 'High Inventory'))
plt.title("Venn Diagram of High Production, High Sales, and High Inventory Months")
plt.show()
```

```python
# Pertanyaan 4: 
# Buat pie chart yang menunjukkan persentase total produk yang diproduksi setiap bulan terhadap total produksi tahunan.

import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Month": ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"],
    "Production": [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600]
}

df = pd.DataFrame(data)

# Calculate total production per month as a percentage of total annual production
total_annual_production = df["Production"].sum()
production_percentage = (df["Production"] / total_annual_production) * 100

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(production_percentage, labels=df["Month"], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Persentase Total Produksi Bulanan terhadap Produksi Tahunan")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Pertanyaan 5:
# Buat bar chart untuk membandingkan jumlah produk yang diproduksi (Production), terjual (Sales), dan dikembalikan (Returns) setiap bulan.

import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Month": ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"],
    "Production": [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    "Sales": [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    "Returns": [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

# Set position of bar on X axis
bar_width = 0.25
r1 = range(len(df["Month"]))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create bar chart
plt.figure(figsize=(14, 8))
plt.bar(r1, df["Production"], color='blue', width=bar_width, edgecolor='grey', label='Production')
plt.bar(r2, df["Sales"], color='green', width=bar_width, edgecolor='grey', label='Sales')
plt.bar(r3, df["Returns"], color='red', width=bar_width, edgecolor='grey', label='Returns')

# Add labels
plt.xlabel('Month', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(df["Month"]))], df["Month"])
plt.ylabel('Number of Products', fontweight='bold')
plt.title('Comparison of Production, Sales, and Returns per Month')

# Add legend
plt.legend()

# Show plot
plt.show()
```
