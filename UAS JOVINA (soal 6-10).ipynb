```python
# Pertanyaan 1: Histogram
# Bagaimana distribusi jumlah produk yang terjual setiap bulan (Sales)?
# Visualisasi: Histogram

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Data dalam bentuk tabel
data = {
    'Month': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Production': [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    'Sales': [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    'Inventory': [100, 200, 300, 200, 100, 200, 100, 200, 300, 300, 200, 100],
    'Returns': [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.hist(df['Sales'], bins=5, edgecolor='black')
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

```python
# Pertanyaan 3: Diagram Venn
# Misalkan bulan-bulan yang jumlah pengembaliannya di atas rata-rata disebut "High Returns", bulan-bulan yang jumlah penjualannya di atas rata-rata disebut "High Sales", dan bulan-bulan yang jumlah produksinya di atas rata-rata disebut "High Production". Buat diagram Venn untuk menggambarkan hubungan antara tiga kelompok bulan ini.
# Visualisasi: Diagram Venn

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Data dalam bentuk tabel
data = {
    'Month': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Production': [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    'Sales': [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    'Inventory': [100, 200, 300, 200, 100, 200, 100, 200, 300, 300, 200, 100],
    'Returns': [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

average_returns = df['Returns'].mean()
average_sales = df['Sales'].mean()
average_production = df['Production'].mean()

high_returns = set(df[df['Returns'] > average_returns]['Month'])
high_sales = set(df[df['Sales'] > average_sales]['Month'])
high_production = set(df[df['Production'] > average_production]['Month'])

plt.figure(figsize=(8, 6))
venn3([high_returns, high_sales, high_production], ('High Returns', 'High Sales', 'High Production'))
plt.title('Venn Diagram of Returns, Sales, and Production')
plt.show()
```

```python
# Pertanyaan 4: Pie Chart
# Bagaimana persentase total produk yang terjual setiap bulan (Sales) terhadap total penjualan tahunan?
# Visualisasi: Pie Chart

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Data dalam bentuk tabel
data = {
    'Month': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Production': [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    'Sales': [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    'Inventory': [100, 200, 300, 200, 100, 200, 100, 200, 300, 300, 200, 100],
    'Returns': [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.pie(df['Sales'], labels=df['Month'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Sales per Month')
plt.axis('equal')
plt.show()
```

```python
# Pertanyaan 5: Bar Chart
# Bagaimana perbandingan jumlah produk yang diproduksi (Production), terjual (Sales), dan dikembalikan (Returns) untuk tiga bulan tertentu (misalnya, Januari, Februari, Maret)?
# Visualisasi: Bar Chart

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Data dalam bentuk tabel
data = {
    'Month': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Production': [1000, 1500, 2000, 1700, 1300, 1800, 1600, 1900, 2100, 2200, 1700, 1600],
    'Sales': [900, 1400, 1900, 1600, 1200, 1700, 1500, 1800, 2000, 2100, 1600, 1500],
    'Inventory': [100, 200, 300, 200, 100, 200, 100, 200, 300, 300, 200, 100],
    'Returns': [50, 60, 70, 80, 50, 40, 30, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

selected_months = df[df['Month'].isin(['Januari', 'Februari', 'Maret'])]
bar_width = 0.25
months = selected_months['Month']

r1 = range(len(selected_months))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.figure(figsize=(10, 6))
plt.bar(r1, selected_months['Production'], color='b', width=bar_width, edgecolor='grey', label='Production')
plt.bar(r2, selected_months['Sales'], color='g', width=bar_width, edgecolor='grey', label='Sales')
plt.bar(r3, selected_months['Returns'], color='r', width=bar_width, edgecolor='grey', label='Returns')

plt.xlabel('Month')
plt.xticks([r + bar_width for r in range(len(selected_months))], selected_months['Month'])
plt.ylabel('Number of Products')
plt.title('Bar Chart of Production, Sales, and Returns for Selected Months')
plt.legend()
plt.show()
```
